```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Configure Selenium to use a web driver
driver = webdriver.Chrome()  # Provide the path to your Chrome driver executable if needed

url = 'https://www.natwest.com/global/commercial-currency-rates.html'

# Load the webpage
driver.get(url)

# Extract the page source after the dynamic content has loaded
html_content = driver.page_source

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table element
table = soup.find('table')

# Extract the table data
table_data = []
headers = []
for row in table.find_all('tr'):
    if not headers:
        # Extract the table headers
        headers = [header.get_text(strip=True) for header in row.find_all('th')]
    else:
        # Extract the table rows
        row_data = [data.get_text(strip=True) for data in row.find_all('td')]
        if row_data:
            table_data.append(row_data)

# Print the headers and table data
print(headers)
for row in table_data:
    print(row)

# Close the Selenium webdriver
driver.quit()
```

```python
import pandas as pd
max_columns = max(len(row) for row in table_data)
headers_extended = headers + [''] * (max_columns - len(headers))
df = pd.DataFrame(table_data, columns=headers_extended )
# Replace values in columns 2 and 3 with values from columns 4 and 5
df.iloc[:, 1] = df.iloc[:, 3]  # Column 2 replaced with values from Column 4
df.iloc[:, 2] = df.iloc[:, 4]  # Column 3 replaced with values from Column 5
# Delete columns 4 and 5
df = df.drop(columns=[headers_extended[3], headers_extended[4]])
df = df.reset_index(drop=True)
df
```

```python
from datetime import date

# Assuming you have the modified DataFrame without the index column (df) as shown in the previous code

# Get today's date
today = date.today()

# Create a new column with today's date
df.insert(0, "Date", today)

# Print the updated DataFrame
df
```
